```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})
df.head(10)
```

Створити візуалізацію, аналогічно рисунку 
 - перші 3 графіки візуалізувати, викорстовуючи значення із df
 - останній це пряма пропорційність

```python
fig, ax = plt.subplots(1, 2, figsize=(20,5))
ax[0].plot(df["a"])
ax[1].plot(df["b"])

fig, ax1 = plt.subplots(1, 2, figsize=(20,5))
ax1[0].plot(df["c"])

x = np.linspace(0, 30, 30)
y = x
plt.plot(x,y)

```

Створити візуалізацію, аналогічно рисунку 

```python
plt.figure(figsize=(11, 5))
plt.plot(df["a"])
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`

```python
plt.figure(figsize=(11, 4))
plt.plot(df["a"])

plt.figure(figsize=(11, 4))
plt.plot(df["b"])

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df["a"])
plt.plot(df["b"])
```

Створити візуалізацію, аналогічно рисунку 
- використовуйте колонки `a` та `b`
- задайте стиль 'darkgrid' за допомогою команди `sns.set_style`

```python
plt.figure(figsize=(10, 6))
sns.set_style("darkgrid")
plt.plot(df["a"])
plt.plot(df["b"])
```

Створити візуалізацію, аналогічно рисунку 
- для колонки `a` використайте червоний колір та лінію формату `-.`

- для колонки `b` використайте помаранчевий колір та товщину `10`

- для колонки `c` використайте жовтий колір та товщину `1` і маркер `o`


```python
plt.figure(figsize=(10, 6))
sns.set_style("darkgrid")


plt.plot(x, df["a"] , color='red', linestyle ='-.')
plt.plot(x, df["b"], color='orange', linewidth=6)
plt.plot(x, df["c"], color='yellow', linewidth=1, marker='o',markersize=4 )
```

Створити візуалізацію, аналогічно рисунку 
- і не забудьте про легенду :)

```python
plt.figure(figsize=(11, 2))
plt.yticks(fontsize=7)
plt.xticks(fontsize=7)
plt.plot(df["a"], linewidth=1, label="Line A")
plt.legend()

plt.figure(figsize=(11, 2))
plt.yticks(fontsize=7)
plt.xticks(fontsize=7)
plt.plot(df["b"], linewidth=1, label="Line B" )
plt.legend()

plt.figure(figsize=(11, 2))
plt.yticks(fontsize=7)
plt.xticks(fontsize=7)
plt.plot(df["c"], linewidth=1)
```

Створити візуалізацію, аналогічно рисунку 
- використайте томатний колір та відстань між стовбцями 0.5

```python
plt.figure(figsize=(11, 5))

x = df.index
y = df["a"]

plt.yticks(fontsize=7)
plt.xticks(fontsize=7)

plt.bar(x, y, width=0.5, label = "a", color = "tomato")
plt.legend()
```

Створити візуалізацію, аналогічно рисунку 
- добавте всі підписи та правильний маркер

```python
plt.figure(figsize=(12, 5))
plt.yticks(fontsize=7)
plt.xticks(fontsize=7)
plt.title("This is the Title", fontsize=7)
plt.xlabel ("This is the X Axis", fontsize=11)
plt.ylabel ("This is the Y Axis", fontsize=7)
plt.grid(True)
plt.xticks(range(0, 30, 1))

plt.plot(df["a"], linewidth=1, marker='^', markersize = 4)
```
